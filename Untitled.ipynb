```python
import pandas as pd
import numpy as np
import sklearn  as 
```

```python
df = pd.read_csv('DataScience_project.csv')
display(df)
```

```python
df = df.drop(["Score"],axis=1)
display(df)
```

```python
#renaming the column-names for easier access
df.rename(columns={'Health Issues(if any)': 'health-problems','Waiting Time (approx)':'WT',
                   'Veg or Non Veg':'veg/non-veg','Avg Amount paid..':'Amount','Spicy..?':'Spicy'}, inplace=True)
          
```

```python
# Data discritization
#bin edges
bin_edges = [0, 50, 100, 150, 200]

#labels for the bins
bin_labels = ['Low', 'Medium', 'High', 'Very High']

#cut() function to perform data binning
df['Payment-Contribution'] = pd.cut(df['Amount'], bins=bin_edges, labels=bin_labels)
```

```python
#Data-encoding
ordinal_mapping = {'Male': 'M', 'Female': 'F', '.': 'G'}
df['Gender'] = df['Gender'].map(ordinal_mapping)
display(df)
```
